```python
import numpy as np
import cv2 as cv
import time
import matplotlib.pyplot as plt
from utils import DLT

# =========================
# Video name
# =========================
Nvideo = 7
VideoName = 'C:\\Users\\Enzo\\Desktop\\suseso_marcha2\\RealSense_Marcha\\Calibration_Videos_RS\\Calib_Video'+ str(Nvideo)

VideoA = VideoName + 'A.avi'
VideoB = VideoName + 'B.avi'
print(VideoA)
print(VideoB)


# =========================
# 1. Define function to calibrate two cameras
# =========================
def calibrate_both_cameras(VideoNameA, VideoNameB):

    capA = cv.VideoCapture(VideoNameA)
    capB = cv.VideoCapture(VideoNameB)

    if not capA.isOpened():
        raise ValueError(f'No se pudo abrir el video A: {VideoNameA}')
    if not capB.isOpened():
        raise ValueError(f'No se pudo abrir el video B: {VideoNameB}')

    find_flags = cv.CALIB_CB_ADAPTIVE_THRESH + cv.CALIB_CB_NORMALIZE_IMAGE
    world_scaling = 30.0   # cambia esto al tamaño real del cuadrado

    # termination criteria
    criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 60, 0.0001)

    # pattern size = esquinas internas
    pattern_size = (8, 6)

    # prepare object points
    objp = np.zeros((pattern_size[0] * pattern_size[1], 3), np.float32)
    objp[:, :2] = np.mgrid[0:pattern_size[0], 0:pattern_size[1]].T.reshape(-1, 2)
    objp = world_scaling * objp

    # Arrays to store object points and image points from all the images.
    objpoints = []   # 3d points in real world space
    imgpointsA = []  # 2d points in image plane
    imgpointsB = []  # 2d points in image plane

    counter = 0
    greyFrameA = None
    greyFrameB = None

    while True:
        retA, frameA = capA.read()
        retB, frameB = capB.read()

        # CORRECCIÓN CLAVE:
        # salir del loop cuando el video se termina
        if not retA or not retB:
            print('Fin del video o no se pudo leer uno de los frames.')
            break

        counter += 1

        # tomar 1 de cada 4 frames, lo cambie a 1 de cada 2
        if np.mod(counter, 3) == 0:

            greyFrameA = cv.cvtColor(frameA, cv.COLOR_BGR2GRAY)
            greyFrameB = cv.cvtColor(frameB, cv.COLOR_BGR2GRAY)

            # Find the chessboard corners
            retA1, cornersA = cv.findChessboardCorners(greyFrameA, pattern_size, flags=find_flags)
            retB1, cornersB = cv.findChessboardCorners(greyFrameB, pattern_size, flags=find_flags)

            if retA1 and retB1:
                objpoints.append(objp)

                corners2A = cv.cornerSubPix(greyFrameA, cornersA, (11, 11), (-1, -1), criteria)
                corners2B = cv.cornerSubPix(greyFrameB, cornersB, (11, 11), (-1, -1), criteria)

                imgpointsA.append(corners2A)
                imgpointsB.append(corners2B)

                # Draw and display the corners
                frameA_draw = frameA.copy()
                frameB_draw = frameB.copy()

                cv.drawChessboardCorners(frameA_draw, pattern_size, corners2A, retA1)
                cv.drawChessboardCorners(frameB_draw, pattern_size, corners2B, retB1)

                frame_cat = cv.hconcat([frameA_draw, frameB_draw])
                half = cv.resize(frame_cat, (0, 0), fx=0.5, fy=0.5)
                cv.imshow('Frame', half)

            # una sola espera de teclado por loop
            if cv.waitKey(1) & 0xFF == ord('q'):
                print('Interrumpido por teclado.')
                break

    capA.release()
    capB.release()
    cv.destroyAllWindows()

    # chequeos de seguridad
    if greyFrameA is None or greyFrameB is None:
        raise ValueError('No se pudo convertir ningún frame a gris. Revisa los videos.')

    if len(objpoints) == 0:
        raise ValueError('No se detectó el checkerboard simultáneamente en ambas cámaras.')

    # Calibrate each camera
    retA_cal, mtxA, distA, rvecsA, tvecsA = cv.calibrateCamera(
        objpoints, imgpointsA, greyFrameA.shape[::-1], None, None
    )

    retB_cal, mtxB, distB, rvecsB, tvecsB = cv.calibrateCamera(
        objpoints, imgpointsB, greyFrameB.shape[::-1], None, None
    )

    return objpoints, imgpointsA, greyFrameA, retA_cal, mtxA, distA, imgpointsB, greyFrameB, retB_cal, mtxB, distB


# =========================
# 2. Detect corners of the chessboard
# =========================
objpoints, imgpointsA, greyFrameA, retA, mtxA, distA, imgpointsB, greyFrameB, retB, mtxB, distB = calibrate_both_cameras(VideoA, VideoB)

print("Number of frames detected:", len(imgpointsA))


# =========================
# 3. Stereo Calibration
# =========================
Time0 = time.time()

stereocalibration_flags = cv.CALIB_FIX_INTRINSIC
criteria_stereo = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 100, 0.00001)

ret3D, CM1, dist1, CM2, dist2, R, T, E, F = cv.stereoCalibrate(
    objpoints,
    imgpointsA,
    imgpointsB,
    mtxA,
    distA,
    mtxB,
    distB,
    greyFrameA.shape[::-1],
    criteria=criteria_stereo,
    flags=stereocalibration_flags
)

duracion = time.time() - Time0
print('El codigo se demoro', duracion, '[s]')


# =========================
# 4. Print calibration values
# =========================
print('retA =', retA)
print('retB =', retB)
print('ret3D =', ret3D)


# =========================
# 5. Projection matrices
# =========================
# RT matrix for C1 is identity
RT1 = np.concatenate([np.eye(3), [[0], [0], [0]]], axis=-1)
P1 = mtxA @ RT1

# RT matrix for C2 is the R and T obtained from stereo calibration
RT2 = np.concatenate([R, T], axis=-1)
P2 = mtxB @ RT2


# =========================
# 6. Save calibration parameters
# =========================
np.savez(
    '.\\Calibration_Videos_RS\\Calibration_parameters_' + str(Nvideo),
    ret3D=ret3D,
    CM1=CM1,
    dist1=dist1,
    CM2=CM2,
    dist2=dist2,
    R=R,
    T=T,
    E=E,
    F=F,
    P1=P1,
    P2=P2
)


# =========================
# 7. Visual test of calibration
# Plot one point in 3D space in both cameras
# =========================
Nframes = len(imgpointsA) - 1

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.imshow(greyFrameA, cmap='gray')
ax1.plot(imgpointsA[Nframes][-1][0][0], imgpointsA[Nframes][-1][0][1], '.r')
ax1.set_title('Camera A')

ax2.imshow(greyFrameB, cmap='gray')
ax2.plot(imgpointsB[Nframes][-1][0][0], imgpointsB[Nframes][-1][0][1], '.r')
ax2.set_title('Camera B')

plt.show()


# =========================
# 8. Control: Plot all corners of the board and check they are on a plane
# =========================
from mpl_toolkits.mplot3d import Axes3D

colours = ['red', 'orange', 'yellow', 'lime', 'cyan', 'blue']

fig2 = plt.figure()
ax = fig2.add_subplot(111, projection='3d')

for iframes in np.arange(0, Nframes):
    D3Points = []

    for i in np.arange(0, 8 * 6):
        D3Point = DLT(P1, P2, imgpointsA[iframes][i][0], imgpointsB[iframes][i][0])
        D3Points.append(D3Point)

    D3Points = np.array(D3Points)

    if iframes == 0:
        Maxs = np.max(D3Points, axis=0) + 10
        Mins = np.min(D3Points, axis=0) - 10

    for i in np.arange(0, 8 * 6 - 1):
        row = int(np.floor(i / 8))
        ax.plot(D3Points[i][0], D3Points[i][1], D3Points[i][2], '.', color=colours[row])
        ax.plot(
            xs=[D3Points[i][0], D3Points[i + 1][0]],
            ys=[D3Points[i][1], D3Points[i + 1][1]],
            zs=[D3Points[i][2], D3Points[i + 1][2]],
            c=colours[row]
        )

    ax.set_xlim3d([Mins[0], Maxs[0]])
    ax.set_ylim3d([Mins[1], Maxs[1]])
    ax.set_zlim3d([Mins[2], Maxs[2]])
    ax.view_init(-41, -67)

    plt.show()
    plt.pause(0.01)
    ax.cla()

print("Resolución calibración:", greyFrameA.shape)
```
